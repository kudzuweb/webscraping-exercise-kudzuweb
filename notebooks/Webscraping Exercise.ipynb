```python
#packages

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#request page & check response

URL = 'https://realpython.github.io/fake-jobs/'

response = requests.get(URL)

response
```

```python
#convert to BeautifulSoup object & print readable HTML

soup = BeautifulSoup(response.text)

print(soup.prettify())
```

```python
# 1a) use .find to to find the tag containing the first job title "Senior Python Developer"

title1 = soup.find('h2').text

title1
```

```python
# 1b) find all titles and turn into a list

htmltitles = soup.findAll('h2')

titles = []

for title in htmltitles:
    titles.append(title.text)

print(titles)
type(titles)
```

```python
# 1c) extract the companies, locations, and posting dates for each job and turn into a clean list

htmlcompanies = soup.findAll('h3')
companies = []

for company in htmlcompanies:
    companies.append(company.text)
```

```python
# 1c) locations

htmllocations = soup.findAll('p', attrs={'location'})
locations = []

for location in htmllocations:
    locations.append(location.text.strip())

locations
```

```python
# 1c) posting dates

htmlposting_dates = soup.findAll('time')

posting_dates = []

for posting_date in htmlposting_dates:
    posting_dates.append(posting_date.text)
    
posting_dates
```

```python
# 1d) create dataframe from lists titles, companies, locations, & posting_dates

job_listings = pd.DataFrame(list(zip(titles, companies, locations, posting_dates)), columns=['Job Titles', 'Companies', 'Locations', 'Date Posted'])
job_listings
```
